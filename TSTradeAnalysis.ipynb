```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Case - Analysis.xlsx')
df
```

```python
df.dropna(inplace=True)
df
```

```python
df1 = df.sort_values(by=['Growth Rate Value(INR)','Growth Rate Value(USD Mn)'],ascending = [False,False])
```

```python
df1 = df1.round(2)
```

```python
sns.set(rc={'figure.figsize':(15,5)})
GRINR = sns.barplot(x='Sector ',y='Growth Rate Value(INR)',data=df1)
for i in GRINR.containers:
    GRINR.bar_label(i)
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(15,5)})
GRINR = sns.barplot(x='Sector ',y='Growth Rate Value(USD Mn)',data=df1)
for i in GRINR.containers:
    GRINR.bar_label(i)
plt.xticks(rotation = 90)
plt.show()
```

```python
df = df.round(2)
```

```python
groth_rate_diff = df['Growth Rate Value(INR)'] - df['Growth Rate Value(USD Mn)']
```

```python
groth_rate_diff
```

```python
df['Difference_in_GR'] = groth_rate_diff
```

```python
df = df.rename(columns = {'Growth Rate Value(INR)':'Growth Rate % Value(INR)','Growth Rate Value(USD Mn)':'Growth Rate % Value(USD Mn)'})
```

```python
df
```

```python
df.columns
```

```python
df['Contribution_in_total_22_23_%'] = (df[' April, 22 To March, 23 Value(INR)']/df[' April, 22 To March, 23 Value(INR)'].sum())*100
```

```python
df
```

```python
df['Contribution_in_total_23_24_%'] = (df[' April, 23 To March, 24 Value(INR)']/df[' April, 23 To March, 24 Value(INR)'].sum())*100
```

```python
df
```

```python
SecPercentbyTotal22 = sns.barplot(x='Sector ',y='Contribution_in_total_22_23_%',data=df.sort_values(by='Contribution_in_total_22_23_%',ascending = False))
for i in SecPercentbyTotal22.containers:
    SecPercentbyTotal22.bar_label(i)
plt.xlabel('Sector')
plt.ylabel('Contribution in total 22-23 in %')
plt.xticks(rotation = 90)
plt.show()
```

```python
df = df.round(2)
```

```python
SecPercentbyTotal23 = sns.barplot(x='Sector ',y='Contribution_in_total_23_24_%',data=df.sort_values(by='Contribution_in_total_23_24_%',ascending = False))
for i in SecPercentbyTotal23.containers:
    SecPercentbyTotal23.bar_label(i)
plt.xlabel('Sector')
plt.ylabel('Contribution in total 23-24 in %')
plt.xticks(rotation = 90)
plt.show()
```
